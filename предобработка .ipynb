## 2.  Проверка ошибок в данных и их предобработка


### 2.1. Названия, или метки, столбцов датафрейма

- Выведем на экран названия всех столбцов датафрейма и проверим их стиль написания.
- Приведем все столбцы к стилю snake case. Названия должны быть в нижнем регистре, а вместо пробелов — подчёркивания.

```python
# выведем названия всех столбцов датафрейма
games.columns
```

```python
# заменим в названиях столбцов пробел на подчеркивание и приведем все названия к нижнему регистру
names = []
for name in games.columns:
    name = name.replace(' ', '_').lower()
    names.append(name)
games.columns = names
```

Мы привели названия столбцов датафрейма к более удобному формату для дальнейшей работы с ними. 

### 2.2. Типы данных

Ранее мы выяснили, что в полях `eu_sales, jp_sales, user_score` стоит изменить тип данных на `float`,  в `year_of_release` на `int`.

```python
''' преобразуем данные в столбцах eu_sales, jp_sales, user_score к типу float, используя параметр downcast, чтобы подобрать самый оптимальный тип 
данных, и параметр coerce, чтобы нечисловые значения заменились на Nan '''
games['eu_sales'] = pd.to_numeric(games['eu_sales'], errors = 'coerce', downcast = 'float')
games['jp_sales'] = pd.to_numeric(games['jp_sales'], errors = 'coerce', downcast = 'float')
games['user_score'] = pd.to_numeric(games['user_score'], errors = 'coerce', downcast = 'float')
```

Также изменим тип данных в столбцах `na_sales`, `other_sales` и `critic_score` с float64 на более оптимальный.

```python
# с помощью параметра downcast подбираем самый оптимальный тип данных
games['na_sales'] = pd.to_numeric(games['na_sales'], downcast = 'float')
games['other_sales'] = pd.to_numeric(games['other_sales'], downcast = 'float')
games['critic_score'] = pd.to_numeric(games['critic_score'], downcast = 'float')
```

```python
# проверим, что у нас получилось
games.info()
```

Заметим, что размер занимаемой памяти значительно уменьшился.

Для оптимизации работы с данными в датафрейме были сделаны следующие изменения типов данных:
- `eu_sales`, `jp_sales`, `user_score`: тип изменен с `object` на `float32`
- `year_of_release`: тип изменен с `float64` на `int16`
- `na_sales`, `other_sales`, `critic_score`: тип изменен с `float64` на `float32`

В столбцах `eu_sales` и `jp_sales` содержится 6 и 4 пропусков соответственно. Они могли появиться из-за технической ошибки при сборе данных, либо из-за отсутсвия продаж в данных странах. Количество пропусков небольшое (примерно 0.035% и 0.024%), поэтому можно заменить их на среднее количество продаж для такой платформы и года выхода игры.

Самое большое количество пропусков в столбцах `critic_score` (8714), `user_score` (9268), `rating` (6871) - примерно 51.4%, 54.7, 40.5% соответственно, что является очень большим количеством. Как мы ранее предположили, это может быть связано с тем, что это все субъективные оценки, и не было найдено какого-то конкретного точного значения, либо с тем, что игра вышла недавно и еще просто не успела собрать достаточное количество оценок. В данном случае мы не будем удалять строки, т.к. они составляют очень большой процент от данных и остальные данные могут быть нам важны. Заменять на среднее значение мы тоже не будем, т.к. это субъективные оценки, и слишком мало данных, на основе которых можно рассчитать среднее, это может исказить результаты анализа. В столбцах `critic_score` и `user_score` заменим пропуски на значение -1, чтобы сохранить числовой тип данных и т.к. оценки не могут быть отрицательными. В столбце `rating` заменим пропуски на текстовый комментарий "Нет данных".

Удалим строки, где нет информации о названии и жанре.

```python
# найдем индексы строк с пропущенным названием
index = games[games['name'].isna()].index
# удалим строки с этими индексами
games = games.drop(index)
```

Заменим пропуски в столбцах `eu_sales` и `jp_sales` на среднее значение для игр с той же платформы и того же года выпуска.

```python
# создадим функцию, которая будет возвращать среднее значение eu_sales для группы игр с той же платформы и того же года выпуска, если значение пропущено 
def change_eu(row):
    if row.isna()['eu_sales']:
        group = games[(games['platform'] == row['platform']) & (games['year_of_release'] == row['year_of_release'])]
        return group['eu_sales'].mean()
    else:
        return row['eu_sales']
# применим функцию ко всем строкам датафрейма
games['eu_sales'] = games.apply(change_eu, axis = 1)
```

```python
# создадим такую же функцию для jp_sales 
def change_jp(row):
    if row.isna()['jp_sales']:
        group = games[(games['platform'] == row['platform']) & (games['year_of_release'] == row['year_of_release'])]
        return group['jp_sales'].mean()
    else:
        return row['jp_sales']
# применим функцию ко всем строкам датафрейма
games['jp_sales'] = games.apply(change_jp, axis = 1)
```

Заменим пропуски в столбцах `critic_score`, `user_score` на -1, а в `rating` на "Нет данных".

```python
# используем метод fillna для заполнения пропусков
games['critic_score'] = games['critic_score'].fillna(-1)
games['user_score'] = games['user_score'].fillna(-1)
games['rating'] = games['rating'].fillna('Нет данных')
```

## 3. Фильтрация данных

Коллеги хотят изучить историю продаж игр в начале XXI века, и их интересует период с `2000 по 2013` год включительно. Отберем данные по этому показателю. Сохраним новый срез данных в отдельном датафрейме `games_actual`.

```python
# проведем фильтрацию по полю year_of_release
games_actual = pd.DataFrame(games[(games['year_of_release'] >= 2000) & (games['year_of_release'] <= 2013)])
# проверим правильность преобразования, выведем минимальный и максимальный год 
print(games_actual['year_of_release'].min())
games_actual['year_of_release'].max()
```

Мы получили датафрейм, в котором хранятся данные только за интересующие нас года.

## 4. Категоризация данных
    
Проведем категоризацию данных:
- Разделим все игры по оценкам пользователей и выделим такие категории: высокая оценка (от 8 до 10 включительно), средняя оценка (от 3 до 8, не включая правую границу интервала) и низкая оценка (от 0 до 3, не включая правую границу интервала).

```python
# создадим функцию для категоризации
def categorize(row):
    if row['user_score'] < 0:
        return 'Нет оценки'
    elif row['user_score'] < 3:
        return 'Низкая оценка'
    elif row['user_score'] < 8:
        return 'Средняя оценка'
    else:
        return 'Высокая оценка'
# создадим новый столбец 'user_score_category' с категорией оценки, применим функцию categorize к каждой строке датафрейма
games_actual['user_score_category'] = games_actual.apply(categorize, axis = 1)
```

- Разделим все игры по оценкам критиков и выделим такие категории: высокая оценка (от 80 до 100 включительно), средняя оценка (от 30 до 80, не включая правую границу интервала) и низкая оценка (от 0 до 30, не включая правую границу интервала).

```python
# создадим функцию для категоризации
def categorize(row):
    if row['critic_score'] < 0:
        return 'Нет оценки'
    elif row['critic_score'] < 30:
        return 'Низкая оценка'
    elif row['critic_score'] < 80:
        return 'Средняя оценка'
    else:
        return 'Высокая оценка'
# создадим новый столбец 'critic_score_category' с категорией оценки, применим функцию categorize к каждой строке датафрейма
games_actual['critic_score_category'] = games_actual.apply(categorize, axis = 1)
```

- После категоризации данных проверим результат: сгруппируем данные по выделенным категориям и посчитайте количество игр в каждой категории.

```python
# сгруппируем данные по категории оценок пользователей и посчитаем сколько игр в каждой категории
user_scores = games_actual.groupby('user_score_category')['name'].count()
user_scores
```

Больше всего игр по оценкам пользователей в категории "нет оценки" и "средняя оценка", меньше всего в категории "низкая оценка".

```python
# сгруппируем данные по категории оценок критиков и посчитаем сколько игр в каждой категории
critic_scores = games_actual.groupby('critic_score_category')['name'].count()
critic_scores
```

Больше всего игр по оценкам критиков также в категории "нет оценки" и "средняя оценка", меньше всего в категории "низкая оценка".

- Выделим топ-7 платформ по количеству игр, выпущенных за весь актуальный период.

```python
# посчиатем сколько игр было выпущено на каждой платформе
top = games_actual.groupby('platform')['name'].count()
# отсортируем по убыванию и выведем первые 7 строк
top.sort_values(ascending = False).head(7)
```

Больше всего игр за 2000-2013 гг. было выпущено на платформе `PS2` (2127 игр). Также близка по количеству выпущенных игр платформа `DS` (2120), на которой вышло на 7 игр меньше, чем на PS2.
